## Step 1: Import librerie

```python
# INPUT
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.metrics import RocCurveDisplay, PrecisionRecallDisplay
import joblib
```

---
## Step 2: Caricamento dataset

```python
# INPUT
df = pd.read_csv("data/creditcard.csv")
df.head()
```

```python
# OUTPUT
df.head()
```

## Step 3: Analisi del dataset

```python
# INPUT
df.info()
df.describe()
```

```python
# OUTPUT
df.info()
df.describe()
```

---
## Step 4: Analisi del target (Class)

```python
# INPUT
df['Class'].value_counts()
```

```python
# OUTPUT
df['Class'].value_counts()
df['Class'].value_counts().plot(kind='bar')
plt.title("Distribuzione Classi (0 = legit, 1 = fraud)")
plt.show()
```

---
## Step 5: Analisi delle feature

```python
# INPUT
df['Amount'].hist(bins=100)
plt.title("Distribuzione Amount")
plt.show()

df['Time'].hist(bins=100)
plt.title("Distribuzione Time")
plt.show()
```

---
## Step 6: Preprocessing

```python
# INPUT
X = df[['Time', 'Amount']]
y = df['Class']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

---
## Step 7: Modello Logistic Regression

```python
# INPUT
log_reg = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
log_reg.fit(X_train, y_train)

y_pred_lr = log_reg.predict(X_test)
y_proba_lr = log_reg.predict_proba(X_test)[:,1]
```

```python
# OUTPUT
confusion_matrix(y_test, y_pred_lr)
classification_report(y_test, y_pred_lr)
print("ROC AUC:", roc_auc_score(y_test, y_proba_lr))

RocCurveDisplay.from_predictions(y_test, y_proba_lr)
plt.show()
PrecisionRecallDisplay.from_predictions(y_test, y_proba_lr)
plt.show()
```

---
## Step 8: Modello Random Forest

```python
# INPUT
rf = RandomForestClassifier(class_weight='balanced', n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
y_proba_rf = rf.predict_proba(X_test)[:,1]
```

```python
# OUTPUT
confusion_matrix(y_test, y_pred_rf)
classification_report(y_test, y_pred_rf)
print("ROC AUC:", roc_auc_score(y_test, y_proba_rf))

RocCurveDisplay.from_predictions(y_test, y_proba_rf)
plt.show()
PrecisionRecallDisplay.from_predictions(y_test, y_proba_rf)
plt.show()
```

---
## Step 9: Salvataggio modello e scaler

```python
# INPUT
joblib.dump(rf, "fraud_model.pkl")
joblib.dump(scaler, "scaler.pkl")
```
